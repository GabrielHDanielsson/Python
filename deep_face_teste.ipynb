```python
#bibliotecas
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
#imagem teste emoção felicidade
plt.imshow(img[:, :,::-1])
```

```python
analise = DeepFace.analyze(imagem1)
analise
```

```python
#analise de todas emocoes na imagem
analise['emotion']
```

```python
#emocao dominante
analise['dominant_emotion']
```
