```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.graph_objects as go
```

```python
bradesco = yf.download("BBDC3.SA",
                       start="2017-01-01",
                       end="2022-01-01",
                       progress="False")

itau = yf.download("ITUB3.SA",
                    start="2017-01-01",
                    end="2022-01-01",
                    progress="False")

banco_do_brasil = yf.download("BBAS3.SA",
                    start="2017-01-01",
                    end="2022-01-01",
                    progress="False")

santander = yf.download("SANB3.SA",
                    start="2017-01-01",
                    end="2022-01-01",
                    progress="False")
```

```python
#grafico de candles 
itau_candles = go.Figure(data=[go.Candlestick(x=itau.index,
                                         open=itau["Open"],
                                         high=itau["High"],
                                         low=itau["Low"],
                                         close=itau["Close"])])
itau_clandes.show()

```

```python
#Comparao das cotaes entre jan/2017 e jan/2022


itau["Close"].plot(figsize=(20,10), label="Ita") 
bradesco["Close"].plot(figsize=(20,10), label = "Bradesco") 
banco_do_brasil["Close"].plot(figsize=(20,10), label = "Banco do Brasil") 
santander["Close"].plot(figsize=(20,10), label = "Santander") 

#ajustando o grafico
plt.ylim(0,60)
plt.yticks(np.arange(0, 60, 10)) 
plt.margins(x=0)
plt.xlabel("Ano")
plt.ylabel("Valor em reais")
plt.legend()
plt.show()
```
